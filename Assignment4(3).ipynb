```python
#getting name
name=[]

try:

    names=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")

    for i in names:

        name.append(i.text)

except NoSuchElementException:#handling no such element exception

    name.append('No details available')

print(len(name))
print(name)
```

```python
#getting data types.
datatypes=[]
try:

    types=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
    
  
    for i in types:

        datatypes.append(i.text)

except NoSuchElementException:#handling no such element exception

    datatypes.append('No details available')

print(len(datatypes))
print(datatypes)
```

```python
tasks=[]
try:

    task=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
    
  
    for i in task:

        tasks.append(i.text)

except NoSuchElementException:#handling no such element exception

    tasks.append('No details available')

print(len(tasks))
print(tasks)
```

```python
attributetypes=[]
try:

    attributetype=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
    
  
    for i in attributetype:

        attributetypes.append(i.text)

except NoSuchElementException:#handling no such element exception

    attributetypes.append('No details available')

print(len(attributetypes))
print(attributetypes)
```

```python
intances=[]
try:

    instance=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
    
  
    for i in instance:

        intances.append(i.text)

except NoSuchElementException:#handling no such element exception

    intances.append('No details available')

print(len(intances))
print(intances)
```

```python
attributes=[]
try:

    attribute=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
    
  
    for i in attribute:

        attributes.append(i.text)

except NoSuchElementException:#handling no such element exception

    attributes.append('No details available')

print(len(attributes))
print(attributes)
```

```python
years=[]
try:

    year=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
    
  
    for i in year:

        years.append(i.text)

except NoSuchElementException:#handling no such element exception

    years.append('No details available')

print(len(years))
print(years)
```
