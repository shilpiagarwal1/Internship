```python
starlist=[]
def getStar():
    try:
      
        stars=driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]")
        
        for i in stars:
            #print(i.text)
            st1=i.text
            st2=st1.split()[0]
            #print(st2)
            starlist.append(st2)
            
    except NoSuchElementException as e:
        print("EXCEPTION OCCURRS::",e)
        starlist.append('-')
```

```python
reviewlist=[]
def getReviews():
    try:
        reviews=driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
        for i in reviews:
        #print(i.text)
            reviewlist.append(i.text)
    except NoSuchElementException as e:
        print("EXCEPTION OCCURRS::",e)
        reviewlist.append('-')
```

```python
nextlinkurl=[]
urls=driver.find_elements_by_xpath("//a[@class='ge-49M']")
for i in urls:
    #print(i.get_attribute("href"))
    nextlinkurl.append(i.get_attribute("href"))
```

```python
# scrapping first page
import time
for i in nextlinkurl:
    #print(i)
    driver.get(i)
    time.sleep(2)
    getStar()
    getReviews()
```

# Normalization

```python
import re
def  clean_text(df, text_field, new_text_field_name):
    df[new_text_field_name] = df[text_field].astype(str).str.lower()
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))  
    # remove numbers
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r"\d+", "", elem))
    
    return df
```

```python
data_clean = clean_text(Ratings, 'Reviews', 'Reviews_clean')
data_clean.head()
```

# Removing Stopwords

```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
data_clean['Reviews_clean'] = data_clean['Reviews_clean'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
data_clean.head()
```

# Stemming

```python
import nltk 
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
data_clean['text_tokens'] = data_clean['Reviews_clean'].apply(lambda x: word_tokenize(x))
data_clean.head()
```

```python
from nltk.stem import PorterStemmer
def word_stemmer(text):
    stem_text = [PorterStemmer().stem(i) for i in text]
    return stem_text
data_clean['text_tokens_stem'] = data_clean['text_tokens'].apply(lambda x: word_stemmer(x))
data_clean.head()
```

# Lemmatization

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
def word_lemmatizer(text):
    lem_text = [WordNetLemmatizer().lemmatize(i) for i in text]
    return lem_text
data_clean['text_tokens_lemma'] = data_clean['text_tokens'].apply(lambda x: word_lemmatizer(x))
data_clean.head()
```

# Part of Speech (POS) tagging and chunking

```python
def word_pos_tagger(text):
    pos_tagged_text = nltk.pos_tag(text)
    return pos_tagged_text
nltk.download('averaged_perceptron_tagger')
data_clean['text_tokens_pos_tagged'] = data_clean['text_tokens'].apply(lambda x: word_pos_tagger(x))
data_clean.head()
```
