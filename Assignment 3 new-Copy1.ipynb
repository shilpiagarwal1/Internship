```python
#Getting product urls and Titles.
from selenium.common.exceptions import StaleElementReferenceException    

ProductTitle_List=[]
try:
    for i in range(0,len(url_list)):
        print(url_list[i])
        driver.get(url_list[i])#Pages Link

        product_url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")#each watch URL
        for i in product_url:
            time.sleep(2)
            p_url=i.get_attribute("href")#getting the Urls of each watches
            ProductTitle_List.append(i.text)#getting title of each watch.
            productUrlList.append(p_url)

   
            
except StaleElementReferenceException as e:
        print("EXCEPTION:::",e)
print(len(productUrlList))
print(len(ProductTitle_List))
```

```python
from selenium.common.exceptions import NoSuchElementException
#Getting prices

starList=[]
pricelist=[]
ratinglist=[]
delieveryList=[]


returnlist=[]
for i in range(0,len(productUrlList)):
        url_=productUrlList[i]
        driver.get(url_)
       
        time.sleep(2)
        try:
            price=driver.find_element_by_xpath("//table[@class='a-lineitem']/tbody/tr[2]/td[2]")
            print(price.text)
            pricelist.append(price.text)
        except NoSuchElementException as e:
            print("EXCEPTION OCCURRS::",e)
            pricelist.append('-')

        ##getting stars
        try:
            stars=driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']").text
            print(stars)
            starList.append(stars)
        except NoSuchElementException as e:
            print("EXCEPTION OCCURRS::",e)
            starList.append('-')
        
        try:
            ##getting a number of ratings
            num_rating=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']").text
            print(num_rating)
            ratinglist.append(num_rating)
        except NoSuchElementException as e:
            print("EXCEPTION OCCURRS::",e)
            ratinglist.append('-')

            
        try:
            ##getting Expected delievery date
            delievery=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']").text
            print(delievery)
            delieveryList.append(delievery)
        except NoSuchElementException as e:
            print("EXCEPTION OCCURRS::",e)
            delieveryList.append('-')
        
       
```

```python
#Getting colors

color_List=[]
productname=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in productname:
    name_string=i.text
    if name_string.find("(")!=-1:
        color=name_string.split("(")
        
        #print(color[1])
        color_=color[1].split(',')
        #print(color_[0])
        color_List.append(color_[0])
    else:
        
        color_List.append("-")

color_List
```

```python
#Getting Mobile name
Mobile_list=[]
productname=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in productname:
    #print(i.text)
    product=i.text
    if(product.find("(")==-1):
        #print("No paranthesis")
        #print(product)
        Mobile_list.append(product)
    else:
        mobile_=product.split("(",1)
       # print(mobile_[0])
        Mobile_list.append(mobile_[0])

Mobile_list
```

```python
#Getting RAM and ROM
RAM_list=[]
ROM_list=[]
memory=driver.find_elements_by_xpath("//li[@class='rgWa7D'][1]")
for i in memory:
    str1=i.text
    #print(str1)
    if(str1.find("|")==-1):
        #NO RAM and ROM
        RAM_list.append("-")
        ROM_list.append("-")
        
    else:
        RAM=str1.split("|")
        #print(RAM)
        RAM_list.append(RAM[0])
        ROM_list.append(RAM[1])
        
print(RAM_list)  
print(ROM_list)
```

```python
#Getting display size and resolution
displaysize_list=[]
display_resolution_list=[]
display_=driver.find_elements_by_xpath("//li[@class='rgWa7D'][2]")
for i in display_:
    display_=i.text
    print(display_)
    displaysize=display_.split(") ")
    print(displaysize[1])
    displaysize_list.append(displaysize[0]+")")
    display_resolution_list.append(displaysize[1])
print(displaysize_list)
print(display_resolution_list)
```

```python
#Getting Primary and Secondary Camera
primary_camera_list=[]
secondary_camera_list=[]
camera=driver.find_elements_by_xpath("//li[@class='rgWa7D'][3]")
for i in camera:
    cam1=i.text
    if(cam1.find('|')==-1):
        #No camera
        primary_camera_list.append("-")
        secondary_camera_list.append("-")
    else:
        primary_camera_list.append(cam1.split("|")[0])
        secondary_camera_list.append(cam1.split("|")[1])
        
print(primary_camera_list)
print(secondary_camera_list)
```

```python
#Getting Prices
pricelist=[]
prices=driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
for i in prices:
    pricelist.append(i.text)
pricelist   
```

```python
#getting Product URL
urllist=[]
urls=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in urls:
    url_=i.get_attribute("href")
    urllist.append(url_)
urllist
```

```python
##Getting Processor
import time
from selenium.common.exceptions import NoSuchElementException                    

processorList=[]
for i in range(0,len(urllist)):
    #print("URL::::",urllist[i])
    driver.get(urllist[i])
    time.sleep(1)
    read_more_btn=driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']")
    read_more_btn.click()
    try:                                                                                     # Extracting Processor Cores from xpath
        core_chk = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[3]/td[1]')
        if core_chk.text != "Processor Core" :
            core_chk = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[2]/td[1]')
            if core_chk.text != "Processor Core" : 
                raise NoSuchElementException
            else :
                cores = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[2]/td[2]/ul/li')
        else :
            cores = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[3]/td[2]/ul/li')
        processorList.append(cores.text)
    except NoSuchElementException:
        processorList.append('-')
    
    
    
    
processorList
```

```python
#Getting Battery Capacity
import time
from selenium.common.exceptions import NoSuchElementException                    

BatteryCapacityList=[]
for i in range(0,len(urllist)):
    #print("URL::::",urllist[i])
    driver.get(urllist[i])
    time.sleep(1)
    read_more_btn=driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']")
    read_more_btn.click()
    try:
        if driver.find_element_by_xpath('//div[@class="_3k-BhJ"][10]/div').text != "Battery & Power Features" :   # Extracting Battery Capacity from xpath
            if driver.find_element_by_xpath('//div[@class="_3k-BhJ"][9]/div').text == "Battery & Power Features" :
                bat_chk = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][9]/table/tbody/tr/td[1]')
                if bat_chk.text != "Battery Capacity" : raise NoSuchElementException
                bat_cap = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][9]/table/tbody/tr/td[2]/ul/li')                
            elif driver.find_element_by_xpath('//div[@class="_3k-BhJ"][8]/div').text == "Battery & Power Features" :     
                bat_chk = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][8]/table/tbody/tr/td[1]')
                if bat_chk.text != "Battery Capacity" : raise NoSuchElementException
                bat_cap = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][8]/table/tbody/tr/td[2]/ul/li')
            else:
                raise NoSuchElementException
        else :
            bat_chk = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][10]/table/tbody/tr/td[1]')
            if bat_chk.text != "Battery Capacity" : raise NoSuchElementException
            bat_cap = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][10]/table/tbody/tr/td[2]/ul/li')                # Extracting Availability from xpath
        BatteryCapacityList.append(bat_cap.text)
    except NoSuchElementException:
        BatteryCapacityList.append('-')
BatteryCapacityList
```

```python
import pandas as pd
Mobile=pd.DataFrame({})
Mobile['MobileName']=Mobile_list
Mobile['Colour']=color_List
Mobile['RAM']=RAM_list
Mobile['ROM']=ROM_list
Mobile['PrimaryCamera']=primary_camera_list
Mobile['SecondaryCamera']=secondary_camera_list
Mobile['DisplaySize']=displaysize_list
Mobile['DisplayResolution']=display_resolution_list
Mobile['Processor Cores']=processorList
Mobile['Battery Capacity']=BatteryCapacityList
Mobile['Product URL']=urllist
Mobile
```

```python
Mobile.to_csv("Mobiledata.csv")
```

```python
#getting specs
OS_List=[]
display_List=[]
processor_List=[]
memory_list=[]
weight_list=[]
dimension_list=[]
graphicprocessor_list=[]
#Getting OS
os=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[3]/td[3]")
for i in os:
    OS_List.append(i.text)
OS_List

#getting display specs
display=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[4]/td[3]")
for i in display:
    display_List.append(i.text)
display_List


#getting processor specs
processor=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[5]/td[3]")
for i in processor :
    processor_List.append(i.text)
processor_List


#getting memory specs
memory=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[6]/td[3]")
for i in memory :
    memory_list.append(i.text)
memory_list

#getting weight specs
weight=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[7]/td[3]")
for i in weight :
    weight_list.append(i.text)
weight_list

#getting dimension specs
dimensions=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[8]/td[3]")
for i in dimensions :
    dimension_list.append(i.text)
dimension_list

#getting graphic processor specs
graphic=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[9]/td[3]")
for i in graphic :
    graphicprocessor_list.append(i.text)
graphicprocessor_list
```
